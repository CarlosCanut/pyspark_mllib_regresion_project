```python
import pandas as pd
from IPython.display import display, HTML
```

Primero extraemos los datos de todos los partidos desde el 2010 hasta el 2021 y los rankings de cada decada:

```python
# partidos
games = []
for year in [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]:
    games.append(pd.read_csv("datasets/atp_matches_"+ str(year) +".csv"))
all_atp_games = pd.concat(games)

# rankings
ranking_atp_10s = pd.read_csv("datasets/atp_rankings_10s.csv")
ranking_atp_20s = pd.read_csv("datasets/atp_rankings_20s.csv")
all_ranking = []
all_ranking.append(ranking_atp_10s)
all_ranking.append(ranking_atp_20s)
all_ranking_atp = pd.concat(all_ranking)
```
